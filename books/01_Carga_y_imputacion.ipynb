### 1. Carga y visualizacion
En esta sección se importa el archivo original de ventas y se revisa una muestra inicial para validar su estructura.

```python
# Cargar el archivo fuente y mostrar una vista previa del dataset.
import pandas as pd

path_file = "../data/ventas.csv"

df_ventas = pd.read_csv(path_file, header=0, sep=",", encoding="utf-8")

df_ventas.head(10)
```

### 2. Información de valores nulos 
Aquí se inspecciona la presencia de datos faltantes mediante visualizaciones que resumen la completitud del dataset.

```python
# Visualizar los valores faltantes para identificar columnas con problemas de completitud.
import missingno as msno
import matplotlib.pyplot as plt

# Mapa de matriz de valores faltantes
msno.matrix(df_ventas)
plt.show()

# Bar chart de completitud
msno.bar(df_ventas)
plt.show()
```

### 3. Imputación basado en mediana y moda
En este bloque se completan los campos faltantes con reglas simples y luego se exporta una versión limpia del archivo.

```python
# Aplicar imputaciones básicas y guardar el resultado para las siguientes etapas del proyecto.
df_ventas["empleado"] = df_ventas["empleado"].fillna(df_ventas["empleado"].mode()[0])
df_ventas["total"] = df_ventas["total"].fillna(df_ventas["total"].median())

print("Valores imputados correctamente.")

# Exportar el CSV limpio
df_ventas.to_csv("../data/ventas_imputado.csv", index=False)
print("CSV guardado correctamente como 'ventas_imputado.csv'.")

msno.matrix(df_ventas)
```
